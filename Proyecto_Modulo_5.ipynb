```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Configuración estética global
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

# Carga de datos del Módulo 4
df = pd.read_csv('dataset_final_ecommerce.csv')
print("✅ Dataset cargado. Dimensiones:", df.shape)
```

```python
# Gráfico 1: Histograma de Edades
plt.figure(figsize=(10, 5))
sns.histplot(df['Edad'], bins=10, kde=True, color='skyblue')
plt.title('Distribución de Edades de los Clientes', fontsize=14)
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()

# Gráfico 2: Boxplot de Monto Total (para ver la dispersión)
plt.figure(figsize=(10, 5))
sns.boxplot(x=df['Monto_Total'], color='lightgreen')
plt.title('Dispersión del Monto Total de Compras', fontsize=14)
plt.xlabel('Monto ($)')
plt.show()
```

Conclusión: La distribución de edades muestra una concentración principal en el rango de los 25 a 45 años. Esto indica que nuestro público objetivo es mayoritariamente adulto joven, lo cual es clave para definir el tono de las campañas de marketing.

Conclusión: El gráfico de caja revela que la mayoría de las compras se sitúan entre los $1,500 y $3,500. Se observa que el tratamiento de outliers realizado en el módulo anterior fue efectivo, manteniendo una dispersión controlada sin valores extremos que distorsionen el promedio.

```python
# Gráfico 3: Scatter Plot (Edad vs Monto Total)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Edad', y='Monto_Total', hue='Categoria_Cliente', style='Categoria_Cliente', s=100)
plt.title('Relación entre Edad y Monto Total de Compras', fontsize=14)
plt.xlabel('Edad (Años)')
plt.ylabel('Monto Total ($)')
plt.legend(title='Categoría')
plt.show()

# Gráfico 4: Bar Plot (Monto Promedio por Ciudad)
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Ciudad', y='Monto_Total', estimator=np.mean, palette='magma', hue='Ciudad', legend=False)
plt.title('Monto Promedio de Gasto por Ciudad', fontsize=14)
plt.xticks(rotation=45)
plt.ylabel('Monto Promedio ($)')
plt.show()
```

Conclusión: No se observa una correlación lineal directa entre la edad y el monto gastado (los puntos están dispersos). Sin embargo, es evidente que los clientes etiquetados como Premium (en color distinto) mantienen consistentemente montos superiores, independientemente de su edad.

Conclusión: Al comparar las ciudades, se observa que Salta y Neuquén presentan los promedios de gasto más altos. Esto sugiere que, aunque tengan menos clientes que Santiago, su valor por cliente es superior, representando una oportunidad de fidelización.

```python
# Gráfico 5: Heatmap (Mapa de Calor) de la Tabla Pivote
# Reutilizamos la lógica de la tabla pivote del módulo anterior
pivot_data = df.pivot_table(values='Monto_Total', index='Rango_Etario', columns='Categoria_Cliente', aggfunc='mean', observed=True)

plt.figure(figsize=(8, 5))
sns.heatmap(pivot_data, annot=True, fmt=".1f", cmap="YlGnBu")
plt.title('Mapa de Calor: Gasto Promedio por Rango Etario y Categoría', fontsize=14)
plt.show()
```

Conclusión: El mapa de calor confirma que el segmento Senior Premium es el que genera el mayor valor por transacción. Por el contrario, los clientes Jóvenes Regulares representan el segmento con menor ticket promedio, lo que sugiere la necesidad de estrategias de upselling.
